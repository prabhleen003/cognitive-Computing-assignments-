```python
#q1
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

r = 7143  
np.random.seed(r)

d = np.random.randint(1000, 5000, size=(12, 4))

m = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df = pd.DataFrame(d, columns=['E', 'C', 'H', 'S'], index=m)

print(df.head())  
print(df.describe())  

t_cat = df.sum()
print(t_cat)

df["T"] = df.sum(axis=1)

df["G"] = df["T"].pct_change() * 100

if r % 2 == 0:
    df["E"] = df["E"] * 0.90  
else:
    df["C"] = df["C"] * 0.85  

plt.figure(figsize=(10, 5))
for col in df.columns[:-2]:  
    plt.plot(df.index, df[col], label=col)
plt.legend()
plt.title("Monthly Sales Trends")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(data=df.iloc[:, :-2])  
plt.title("Sales Distribution")
plt.show()
```

```python
a = np.array([[1, -2, 3], [-4, 5, -6]])

abs_v = np.abs(a)

p_f = np.percentile(a, [25, 50, 75])
p_c = np.percentile(a, [25, 50, 75], axis=0)
p_r = np.percentile(a, [25, 50, 75], axis=1)

s_f = (np.mean(a), np.median(a), np.std(a))
s_c = (np.mean(a, axis=0), np.median(a, axis=0), np.std(a, axis=0))
s_r = (np.mean(a, axis=1), np.median(a, axis=1), np.std(a, axis=1))

print(abs_v, p_f, p_c, p_r)
print(s_f, s_c, s_r)
```

```python
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

floor_vals = np.floor(a)
ceil_vals = np.ceil(a)
trunc_vals = np.trunc(a)
rounded_vals = np.round(a)

print(floor_vals, ceil_vals, trunc_vals, rounded_vals)
```
