```python
import numpy as np
import matplotlib.pyplot as plt

a = float(input("Enter a value for M: "))
b = np.linspace(-10, 10, 400)
c = a * b**2
d = a * np.sin(b)

plt.figure(figsize=(8, 6))
plt.plot(b, c, label=f'y = a*b^2', color='blue', linestyle='-', linewidth=2)
plt.plot(b, d, label=f'y = a*sin(b)', color='green', linestyle='--', linewidth=2)

plt.legend()
plt.grid(True)
plt.title(f"Plot of y = a*b^2 and y = a*sin(b) for a = {a}")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

f = {'Subject': ['Math', 'Science', 'English', 'History', 'Art'],
     'Score': [90, 80, 85, 70, 95]}

g = pd.DataFrame(f)

plt.figure(figsize=(8, 6))
sns.barplot(x='Subject', y='Score', data=g, palette='Set2')

for h, i in enumerate(g['Score']):
    plt.text(h, i + 1, str(i), ha='center', color='black')

plt.title('Scores of Five Subjects')
plt.xlabel('Subject')
plt.ylabel('Score')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

j = 123456
np.random.seed(j)

k = np.random.randn(50)

l, m = plt.subplots(2, 2, figsize=(12, 10))

m[0, 0].plot(np.cumsum(k), label='Cumulative Sum', color='orange')
m[0, 0].set_title('Cumulative Sum Plot')
m[0, 0].set_xlabel('Index')
m[0, 0].set_ylabel('Cumulative Sum')
m[0, 0].grid(True)

m[0, 1].scatter(np.arange(50), k, color='purple')
m[0, 1].set_title('Scatter Plot with Random Noise')
m[0, 1].set_xlabel('Index')
m[0, 1].set_ylabel('Value')
m[0, 1].grid(True)

m[1, 0].plot(k, label='Random Values', color='green')
m[1, 0].set_title('Random Values Plot')
m[1, 0].set_xlabel('Index')
m[1, 0].set_ylabel('Value')
m[1, 0].grid(True)

m[1, 1].hist(k, bins=10, color='blue', edgecolor='black')
m[1, 1].set_title('Histogram of Random Values')
m[1, 1].set_xlabel('Value')
m[1, 1].set_ylabel('Frequency')
m[1, 1].grid(True)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
url = "https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv?raw=true"
o = pd.read_csv(url)
print(o.columns)
plt.figure(figsize=(10, 6))
sns.lineplot(x='month_number', y='total_profit', data=o)
plt.title('Total Profit of All Months')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()
plt.figure(figsize=(10, 6))
sns.lineplot(data=o.drop(columns=['month_number', 'total_units', 'total_profit']), palette="tab10")
plt.title('Sales Data of All Products')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
plt.figure(figsize=(10, 6))
o.drop(columns=['month_number']).plot(kind='bar', figsize=(10, 6))
plt.title('Bar Chart for All Features')
plt.xlabel('Month')
plt.ylabel('Sales/Profit')
plt.grid(True)
plt.show()

```
