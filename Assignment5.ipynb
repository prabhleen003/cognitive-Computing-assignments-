```python
# Q3
i1, i2 = ord('A'), ord('B')
X = i1 + i2
sales = np.array([X, X+50, X+100, X+150, X+200])
```

```python
# b
rate = ((X % 5) + 5) / 100
print(sales * rate)
```

```python
# c
discounted = np.where(sales < X+100, sales * 0.95, sales * 0.90)
print(discounted)
```

```python
# d
weeks = np.vstack([sales, sales * 1.02, sales * 1.04])
print(weeks)
```

```python
# Q4
x = np.linspace(-10, 10, 100)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, x**2, label="x^2")
plt.title("Quadratic Function")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(x, np.exp(x), label='e^x')
```

```python
plt.plot(x, np.log(np.abs(x) + 1), label='log(|x|+1)')
```
